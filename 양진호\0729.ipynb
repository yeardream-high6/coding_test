#지능형 기차
https://www.acmicpc.net/problem/2455

```python
sum=0
people =[]
for i in range(4):    
    a,b = map(int,input().split())        
    sum = (sum-a)+b
    people.append(sum)
print(max(people))
```

#검증수
https://www.acmicpc.net/problem/2475

```python
sr = list(map(int,input().split()))
for i in range(len(sr)):
    sr[i] = sr[i]**2
    
print(sum(sr)%10)
```

#제로
https://www.acmicpc.net/problem/10773

```python
k = int(input())
numlist=[]
for i in range(k):
    zero = int(input())
    if zero == 0:
        numlist.pop()
    else:
        numlist.append(zero)
print(sum(numlist))
```

#상수
https://www.acmicpc.net/problem/2908

```python
a,b =input().split()
if int(a[::-1]) > int(b[::-1]):
    print(int(a[::-1]))
else:
    print(int(b[::-1]))
```
